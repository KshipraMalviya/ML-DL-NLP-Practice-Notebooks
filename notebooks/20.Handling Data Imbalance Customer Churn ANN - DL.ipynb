```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("D:/DataML/churn.csv")
df
```

```python
df=df.drop(['CustomerId','Surname'],axis='columns')
df
```

```python
df=df.drop('RowNumber',axis='columns')
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['Gender'].replace({'Female':1,'Male':0},inplace=True)
df
```

```python
df['Geography'].unique()
```

```python
dummies=pd.get_dummies(df['Geography'], drop_first=True).astype(int)
```

```python
df=pd.concat([df,dummies],axis='columns')
df
```

```python
df.drop('Geography',axis='columns', inplace=True)
df
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
cols_to_scale=['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']
```

```python
scaler=MinMaxScaler()
df[cols_to_scale]=scaler.fit_transform(df[cols_to_scale])
df
```

```python
df.dtypes
```

```python
df['Exited'].value_counts()
```

```python
X=df.drop('Exited', axis='columns')
y=df['Exited']
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(sampling_strategy='minority')
X_sm,y_sm=smote.fit_resample(X,y)
```

```python
y_sm.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_sm,y_sm,test_size=0.2,random_state=18,stratify=y_sm)
```

```python
y_train.value_counts()
```

```python
X
```

```python
X_train.shape
```

```python
model=keras.Sequential([
    keras.layers.Dense(11,input_shape=(11,),activation='relu'),
    keras.layers.Dense(8,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
from sklearn.metrics import classification_report
y_pred=model.predict(X_test)
y_pred_final=[]
for x in y_pred:
    if x>=0.5:
        y_pred_final.append(1)
    else:
        y_pred_final.append(0)
y_pred_final[:5]
```

```python
y_test[0:5]
```

```python
print(classification_report(y_test,y_pred_final))
```
